```python
import cv2
import os

# Root folder of your dataset (update this path)
root_folder = "/content/P_PRUEBA_5-1"   # Example: /content/dental-caries-1

# Output folder where resized images will be saved
output_root = "/content/drive/MyDrive/teeth_resized"

# Target image size
width, height = 610, 540

# Dataset splits
splits = ['train', 'valid', 'test']

# Loop through each split
for split in splits:
    input_folder = os.path.join(root_folder, split, "images")
    output_folder = os.path.join(output_root, split, "images")

    # Create output directory if not exists
    os.makedirs(output_folder, exist_ok=True)

    print(f"\nğŸ“‚ Processing {split} images...")

    for filename in os.listdir(input_folder):
        if filename.lower().endswith(('.jpg', '.jpeg', '.png')):
            img_path = os.path.join(input_folder, filename)
            img = cv2.imread(img_path)

            if img is None:
                print(f"âŒ Skipped: {filename} (cannot open image)")
                continue

            # Resize image
            resized_img = cv2.resize(img, (width, height))

            # Save to output folder
            cv2.imwrite(os.path.join(output_folder, filename), resized_img)

            print(f"âœ… Resized: {filename}")

print("\nğŸ¯ All train, test, and valid images resized to 610x540 successfully!")
```

```python
import cv2
import numpy as np
import os

input_folder = "/content/drive/MyDrive/teeth_resized/train/images"   # example path
output_folder = "/content/drive/MyDrive/teeth_resized/teeth_normalized/train/images"

os.makedirs(output_folder, exist_ok=True)

for filename in os.listdir(input_folder):
    if filename.lower().endswith(('.jpg', '.jpeg', '.png')):
        img_path = os.path.join(input_folder, filename)
        img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)  # read as grayscale
        if img is None:
            continue

        # Normalize pixel values to 0â€“1 range
        normalized_img = img.astype(np.float32) / 255.0

        # Optionally, convert back to 0â€“255 and save as image
        normalized_img_255 = (normalized_img * 255).astype(np.uint8)
        cv2.imwrite(os.path.join(output_folder, filename), normalized_img_255)

print("âœ… All images normalized successfully!")
```
